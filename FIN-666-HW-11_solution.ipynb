## Part I: Problems

1. We discussed two methods that are used for feature representation for text data. What are those methods and what are the main advantages and disadvantages of each one?

The two main methods used for feature representation for text data are:

## 1. Bag of Words (BoW)
Advantages:

Simple and intuitive implementation

Effective for many text classification tasks

Preserves word frequency information

Computationally efficient to implement

Works well with many traditional machine learning algorithms

Disadvantages:

Ignores word order and sentence structure completely

Creates very high-dimensional, sparse matrices (many zeros)

Common words that appear in many documents don't serve as distinguishing features

Cannot capture semantic meaning or context

No notion of word similarity (each word is treated as a completely independent feature)

## 2. Term Frequency-Inverse Document Frequency (TF-IDF)

Advantages:

Addresses the limitation of BoW by weighting terms based on their importance

Downscales words that appear frequently across many documents

Highlights words that are more informative and distinctive to specific documents

Balances local importance (term frequency) with global importance (inverse document frequency)

Performs better than BoW for many information retrieval and text classification tasks

Disadvantages:

Still creates sparse matrices with high dimensionality

Like BoW, ignores word order and semantic relationships

Cannot capture context or word meaning

May give too much weight to rare terms that appear in few documents

Computationally more intensive than simple BoW

Still struggles with synonyms or related words (doesn't recognize semantic similarity)

## Python Implementation (Part a)

```python
# Basic packages:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ML and NLP packages
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, ConfusionMatrixDisplay

import nltk
import string
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

# Load data
df = pd.read_csv('HW11_News_Category_HuffPost_2012_2018_50k_cleaned.csv', encoding='utf-8')
print(f"Dataset shape: {df.shape}")
print(f"Number of categories: {df['category'].nunique()}")

# Split data into training and test samples (80% training, 20% test)
# Using short_description as the main text feature as required in the instructions
y = df['category']
text = df['short_description']  # Using ONLY short_description as specified
text_train, text_test, y_train, y_test = train_test_split(text, y, test_size=0.2, random_state=101)

# Display sample of test data
print("Sample of test data:")
print(text_test.sample(3))

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Define function for stemming
def tokenize(text):
    tokens = nltk.word_tokenize(text)
    tokens = [tkn for tkn in tokens if tkn not in string.punctuation]
    stems = map(PorterStemmer().stem, tokens)
    return stems

# Model 1: Bag of Words (baseline)
print("\nModel 1: Bag of Words (baseline)")
bw = CountVectorizer(lowercase=True)
bw.fit(text_train)
dtm_train = bw.transform(text_train)
dtm_test = bw.transform(text_test)

nb = MultinomialNB()
nb.fit(dtm_train, y_train)
y_pred = nb.predict(dtm_test)

print(f'Accuracy Score = {accuracy_score(y_test, y_pred):.2%}')
print('Classification Report:')
print(classification_report(y_test, y_pred))

# Model 2: Bag of Words with stop words removed
print("\nModel 2: Bag of Words with stop words removed")
bw_stop = CountVectorizer(lowercase=True, stop_words='english')
bw_stop.fit(text_train)
dtm_train = bw_stop.transform(text_train)
dtm_test = bw_stop.transform(text_test)

nb = MultinomialNB()
nb.fit(dtm_train, y_train)
y_pred = nb.predict(dtm_test)

print(f'Accuracy Score = {accuracy_score(y_test, y_pred):.2%}')
print('Classification Report:')
print(classification_report(y_test, y_pred))

# Model 3: Bag of Words with stop words and n-grams (1,2)
print("\nModel 3: Bag of Words with stop words and n-grams (1,2)")
bw_ngram = CountVectorizer(lowercase=True, stop_words='english', ngram_range=(1, 2))
bw_ngram.fit(text_train)
dtm_train = bw_ngram.transform(text_train)
dtm_test = bw_ngram.transform(text_test)

nb = MultinomialNB()
nb.fit(dtm_train, y_train)
y_pred = nb.predict(dtm_test)

print(f'Accuracy Score = {accuracy_score(y_test, y_pred):.2%}')
print('Classification Report:')
print(classification_report(y_test, y_pred))

# Model 4: Bag of Words with stop words, n-grams (1,2) and stemming
print("\nModel 4: Bag of Words with stop words, n-grams (1,2) and stemming")
bw_stem = CountVectorizer(lowercase=True, stop_words='english', ngram_range=(1, 2), tokenizer=tokenize)
bw_stem.fit(text_train)
dtm_train = bw_stem.transform(text_train)
dtm_test = bw_stem.transform(text_test)

nb = MultinomialNB()
nb.fit(dtm_train, y_train)
y_pred = nb.predict(dtm_test)

print(f'Accuracy Score = {accuracy_score(y_test, y_pred):.2%}')
print('Classification Report:')
print(classification_report(y_test, y_pred))

# Compare all models
models = ['BoW Baseline', 'BoW + StopWords', 'BoW + StopWords + N-grams(1,2)', 'BoW + StopWords + N-grams(1,2) + Stemming']
accuracies = []

# Re-run each model briefly to collect accuracies for comparison
# Model 1
bw = CountVectorizer(lowercase=True)
bw.fit(text_train)
dtm_train = bw.transform(text_train)
dtm_test = bw.transform(text_test)
nb = MultinomialNB()
nb.fit(dtm_train, y_train)
y_pred = nb.predict(dtm_test)
accuracies.append(accuracy_score(y_test, y_pred))

# Model 2
bw_stop = CountVectorizer(lowercase=True, stop_words='english')
bw_stop.fit(text_train)
dtm_train = bw_stop.transform(text_train)
dtm_test = bw_stop.transform(text_test)
nb = MultinomialNB()
nb.fit(dtm_train, y_train)
y_pred = nb.predict(dtm_test)
accuracies.append(accuracy_score(y_test, y_pred))

# Model 3
bw_ngram = CountVectorizer(lowercase=True, stop_words='english', ngram_range=(1, 2))
bw_ngram.fit(text_train)
dtm_train = bw_ngram.transform(text_train)
dtm_test = bw_ngram.transform(text_test)
nb = MultinomialNB()
nb.fit(dtm_train, y_train)
y_pred = nb.predict(dtm_test)
accuracies.append(accuracy_score(y_test, y_pred))

# Model 4
bw_stem = CountVectorizer(lowercase=True, stop_words='english', ngram_range=(1, 2), tokenizer=tokenize)
bw_stem.fit(text_train)
dtm_train = bw_stem.transform(text_train)
dtm_test = bw_stem.transform(text_test)
nb = MultinomialNB()
nb.fit(dtm_train, y_train)
y_pred = nb.predict(dtm_test)
accuracies.append(accuracy_score(y_test, y_pred))

# Plot comparison of model accuracies
plt.figure(figsize=(12, 6))
plt.bar(models, accuracies)
plt.title('Model Accuracy Comparison')
plt.ylabel('Accuracy')
plt.xlabel('Model')
plt.ylim(0, 1)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Find the best model
best_model_index = np.argmax(accuracies)
print(f"\nBest performing model: {models[best_model_index]} with accuracy {accuracies[best_model_index]:.2%}")

# Display confusion matrix for best model
if best_model_index == 0:
    best_vectorizer = CountVectorizer(lowercase=True)
elif best_model_index == 1:
    best_vectorizer = CountVectorizer(lowercase=True, stop_words='english')
elif best_model_index == 2:
    best_vectorizer = CountVectorizer(lowercase=True, stop_words='english', ngram_range=(1, 2))
else:
    best_vectorizer = CountVectorizer(lowercase=True, stop_words='english', ngram_range=(1, 2), tokenizer=tokenize)

# Re-fit the best model
best_vectorizer.fit(text_train)
dtm_train = best_vectorizer.transform(text_train)
dtm_test = best_vectorizer.transform(text_test)
best_nb = MultinomialNB()
best_nb.fit(dtm_train, y_train)
y_pred = best_nb.predict(dtm_test)

# Plot detailed confusion matrix for best model
plt.figure(figsize=(14, 12))
ConfusionMatrixDisplay.from_estimator(best_nb, dtm_test, y_test, xticks_rotation=90)
plt.title(f'Confusion Matrix - Best Model: {models[best_model_index]}')
plt.tight_layout()
plt.show()
```

## Python Implementation (Part c)

```python

# Combine headline and short_description as required
# Using Python's string capabilities to combine variables as mentioned in the task
df['combined_text'] = df['short_description'] + " " + df['headline']

# Split data into training and test samples (80% training, 20% test)
y = df['category']
text = df['combined_text']  # Using the combined text
text_train, text_test, y_train, y_test = train_test_split(text, y, test_size=0.2, random_state=101)

# Display sample of test data
print("Sample of combined text data:")
print(text_test.sample(3))

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Apply the best feature representation method found from parts a and b
# The PDF shows TF-IDF with stop words and n-grams (1,2) was the best from part b
print("\nUsing the best model from part b: TF-IDF with stop words and n-grams (1,2)")
best_vectorizer = TfidfVectorizer(lowercase=True, stop_words='english', ngram_range=(1, 2))
best_vectorizer.fit(text_train)
dtm_train = best_vectorizer.transform(text_train)
dtm_test = best_vectorizer.transform(text_test)

# Train Naive Bayes model
nb = MultinomialNB()
nb.fit(dtm_train, y_train)
y_pred = nb.predict(dtm_test)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy Score = {accuracy:.2%}')
print('Classification Report:')
print(classification_report(y_test, y_pred))

# Plot confusion matrix
plt.figure(figsize=(14, 12))
ConfusionMatrixDisplay.from_estimator(nb, dtm_test, y_test, xticks_rotation=90)
plt.title('Confusion Matrix - Combined Text Model (headline + short_description)')
plt.tight_layout()
plt.show()

# Compare with the best model from part b (which used only short_description)
print("\nComparing with best model from part b (short_description only):")
text_only_short = df['short_description']
text_train_short, text_test_short, y_train, y_test = train_test_split(
    text_only_short, y, test_size=0.2, random_state=101)

best_vectorizer_short = TfidfVectorizer(lowercase=True, stop_words='english', ngram_range=(1, 2))
best_vectorizer_short.fit(text_train_short)
dtm_train_short = best_vectorizer_short.transform(text_train_short)
dtm_test_short = best_vectorizer_short.transform(text_test_short)

nb_short = MultinomialNB()
nb_short.fit(dtm_train_short, y_train)
y_pred_short = nb_short.predict(dtm_test_short)
accuracy_short = accuracy_score(y_test, y_pred_short)

print(f'Accuracy with short_description only = {accuracy_short:.2%}')
print(f'Accuracy with combined text = {accuracy:.2%}')
print(f'Improvement by including headline = {(accuracy - accuracy_short):.2%}')

# Bar plot to show improvement
plt.figure(figsize=(10, 6))
models = ['Short Description Only', 'Combined (Headline + Short Description)']
accuracies = [accuracy_short, accuracy]
plt.bar(models, accuracies)
plt.title('Model Accuracy Comparison')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.tight_layout()
plt.show()

print("\nConclusion:")
if accuracy > accuracy_short:
    print(f"Including the headline improved the model's accuracy by {(accuracy - accuracy_short):.2%} percentage points.")
    print("This suggests that headlines contain additional information that helps in categorizing news articles.")
else:
    print("Including the headline did not improve the model's accuracy.")
    print("This suggests that the short descriptions alone contain sufficient information for categorization.")
```

## Part-A, Part-B, and Part-C Analysis Interpretation:

### The textual analysis conducted on the HuffPost news article dataset demonstrates a progressive improvement in classification accuracy through three distinct modeling approaches.
### In Part A, the Bag of Words method was evaluated with various preprocessing techniques. The baseline model achieved 65.05% accuracy, while removing stop words improved this to 66.94%. The most effective configuration incorporated both stop word removal and n-grams (word pairs), reaching 68.34% accuracy. Interestingly, adding stemming slightly decreased performance to 67.39%, suggesting that reducing words to their root forms may have eliminated some distinguishing features between categories.
### Part B applied the more sophisticated TF-IDF method, which weights terms based on their importance across documents. This approach consistently outperformed the Bag of Words models at every level. The baseline TF-IDF model achieved 66.01% accuracy, with the stop word removal configuration reaching 67.98%. Again, the combination of stop word removal and n-grams proved most effective, achieving 69.47% accuracy. As with Bag of Words, adding stemming showed a slight reduction in performance to 68.84%.
### The most dramatic improvement came in Part C, which combined both the headline and short description text using the best configuration from Part B (TF-IDF with stop words and n-grams). This model achieved a remarkable 82.62% accuracy, representing a substantial 13.15 percentage point improvement over using short descriptions alone. This significant jump demonstrates that headlines contain valuable complementary information for categorizing news articles, with each text field providing distinct contextual clues that enhance classification performance.
### The confusion matrices revealed varying classification performance across categories. Sports articles were consistently classified with high accuracy (improving from 81% to 96% recall when headlines were added), while entertainment articles proved more challenging to categorize. These findings highlight the importance of considering multiple text fields when building news categorization systems and confirm TF-IDF with appropriate preprocessing as a strong approach for textual classification tasks.
